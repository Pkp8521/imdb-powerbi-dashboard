```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 4: Check for Missing Values
df.isnull().sum()
```

```python
# üßæ Step 6: Remove Duplicate Rows
df.drop_duplicates(inplace=True)
```

```python
# ‚úÇÔ∏è Step 7: Clean Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
# üß™ Step 8: Fix Data Types
df['year'] = df['year'].astype(int)
df['runtime_(minutes)'] = df['runtime_(minutes)'].astype(int)
```

```python
# üî† Step 9: Clean String Columns
df['genre'] = df['genre'].str.strip().str.lower()
df['director'] = df['director'].str.strip()
```

```python
 # Top 5 highest revenue movies
df.sort_values(by='revenue_(millions)', ascending=False).head()
```

```python
import matplotlib.pyplot as plt
df['year'].value_counts().sort_index().plot(kind='bar', title='Movies per Year')
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

```python
# üíæ Step 10: Save Cleaned Data
df.to_csv('IMDB_Movie_Data_Cleaned.csv', index=False)
```
